```python
short_window = 21
long_window = 34

signals = pd.DataFrame(index=df3.index)
signals['signal'] = 0.0

#SMA of Short Window
signals['short_mavg'] = df3['Close Price'].rolling(window=short_window, min_periods=1,center=False).mean()

#SMA of Long Window
signals['long_mavg'] = df3['Close Price'].rolling(window=long_window,min_periods=1, center=False).mean()

#Create signals
signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] > signals['long_mavg'][short_window:], 1.0,0.0)

signals['positions'] = signals['signal'].diff()
```

```python
fig = plt.figure(figsize=(20,15))

ax1 = fig.add_subplot(111, ylabel='Price')

df3['Close Price'].plot(ax=ax1, color='black', lw=2.)

signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)

ax1.plot(signals.loc[signals.positions == 1.0].index, signals.short_mavg[signals.positions == 1.0], '^' , markersize=20,color='g')


ax1.plot(signals.loc[signals.positions == -1.0].index, signals.short_mavg[signals.positions == -1.0], 'v' , markersize=20,color='r')

plt.show()
```

```python
symbol = 'SUZLON'


df3 = pd.read_csv("G:\harsh\ml\practice problems\intel\week2\dataset\SUZLON.csv".format(symbol), index_col='Date',
                 parse_dates=True, usecols=['Date', 'Close Price'],
                 na_values='nan')

df3 = df3.rename(columns={'Close Price': symbol})
df3.dropna(inplace=True)
sma = df3.rolling(window=14).mean()

rstd = df3.rolling(window=14).std()

upper_band = sma + 2 * rstd
upper_band = upper_band.rename(columns={symbol: 'upper'})
lower_band = sma - 2 * rstd
lower_band = lower_band.rename(columns={symbol: 'lower'})
df3 = df3.join(upper_band).join(lower_band)
ax = df3.plot(title='{} Price and BB'.format(symbol))
ax.fill_between(df3.index, lower_band['lower'], upper_band['upper'], color='#ADCCFF', alpha='0.4')
ax.set_xlabel('Date')
ax.set_ylabel('SMA and BB')
ax.grid()
plt.show()
```
